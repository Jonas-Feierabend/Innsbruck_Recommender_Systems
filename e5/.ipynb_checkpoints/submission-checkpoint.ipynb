Question: "Explain the prediction model in
terms of global average, user bias, item bias, and latent user/item factors. Then identify the
most important hyperparameters, e.g., number of latent dimensions, number of iterations,
regularization, and bias-related parameters."

**The Model**   
The prediction model uses the formula:  
$\hat{r}_{ui}​=μ+b_u​+b_i​+p_u^⊤​q_i$   
where $\hat{r}_{ui}$ is the prediction for one user $u$ and item $i$. 

The components are:  
- $μ$: is the global rating of all films
- $b_u$: is the user bias, because some users rate higher/lower than the average
- $b_i$: the item bias
- $p_u$ is the user-vector and $q_i$ is the itemvector. This factor incorporates the preferences of the user.


**Imporant Hyperparameter**   
The features or number of latent factors are the number of dimensions meaning the length of vectors $p_u, q_i$. They control the complexity of the model, for high features the model can find complex patterns but is more prown to overfit the data. 
For less the model tends to underfit the data. 

The number of iterations control how long the model trains, higher numbers usually lead to better accuracy but need more computational power. 

Regularisiation punishes the model for high weight-values this reduces overfitting but can prevent the model from assuming more complex relationships. 

Damping reduces instability in training, it can help to generalise or to reduce noise. 

## b) 

````python
from lenskit import crossfold as xf
from lenskit.algorithms import als, Recommender
import numpy as np
from sklearn.metrics import mean_squared_error

splits = xf.partition_users(ratings, 1, xf.SampleFrac(0.2))
train, test = next(splits)

grid_features = [10, 50, 100,200]
grid_reg = [0.01, 0.05, 0.2]

results = []


for f in grid_features:
    for r in grid_reg:
        
        mf = als.BiasedMF(
            features=f,
            iterations=20,
            reg=r,
            damping=5
        )
        
        mf.fit(train)

        rec = Recommender.adapt(mf)

        preds = rec.predict(test)


        eval_df = test.copy()
        eval_df["pred"] = preds

        eval_df = eval_df.dropna()


        rmse = np.sqrt(
            mean_squared_error(
                eval_df["rating"],
                eval_df["pred"]
            )
        )

        results.append((f, r, rmse))
        print(f"features={f}, reg={r}, RMSE={rmse:.4f}")


results = sorted(results, key=lambda x: x[2])

print("\nBest configuration:")
print(results[0])
````

````python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame(results, columns=["features", "reg", "rmse"])

grid = df.pivot(index="features", columns="reg", values="rmse")

plt.figure(figsize=(6, 4))
plt.imshow(np.log(grid.values), aspect="auto")

plt.colorbar(label="RMSE")

plt.xticks(
    range(len(grid.columns)),
    grid.columns
)

plt.yticks(
    range(len(grid.index)),
    grid.index
)

plt.xlabel("Regularization (reg)")
plt.ylabel("Latent factors (features)")
plt.title("Matrix Factorization Hyperparameter Grid (RMSE)")

plt.show()
````

As we can see in the Hyperparameter grid, generally the results get better for higher latent factors and for regularization (with exceptions). 
For lower regularization we can see that the amount of latent factors has a huge impact on the performance, interestingely the trend is reversed 
for 10 and 50 latent features and it gets worse first but after that the results improve again. We can see the same 
for a regularization of 0.05. But if we set the regularization to 0.2 it doesnt really matter how high we are setting the latent factors. 
This is a good indication that more features are generally a good thing if you can really fine-tune the parameters. But its better to have a more regulated model beause it can generalise much better. 
This means if raw performance is not the main goal but also efficiency is important, selecting only 10 features with a regularization is enough for a good result. 

## c) 

````python
from lenskit import crossfold as xf
from lenskit.algorithms import als
from lenskit.algorithms import Recommender
from lenskit import topn
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error



splits = xf.partition_users(ratings, 5, xf.SampleFrac(0.2))  # 5-Fold CV

rmse_list = []
mae_list = []
all_metrics = []

rla = topn.RecListAnalysis()
rla.add_metric(topn.ndcg)
rla.add_metric(topn.precision)
rla.add_metric(topn.recall)



for i, (train, test) in enumerate(splits):

    print(f"\n--- Fold {i+1} ---")

  
    mf = als.BiasedMF(
        features=50,
        iterations=20,
        reg=0.05,
        damping=5
    )

    mf.fit(train)
    rec = Recommender.adapt(mf)


    preds = rec.predict(test)

    eval_df = test.copy()
    eval_df["prediction"] = preds
    eval_df = eval_df.dropna()

    rmse = np.sqrt(mean_squared_error(eval_df["rating"], eval_df["prediction"]))
    mae = mean_absolute_error(eval_df["rating"], eval_df["prediction"])

    rmse_list.append(rmse)
    mae_list.append(mae)

    print("RMSE:", rmse)
    print("MAE:", mae)


    all_items = ratings["item"].unique()
    users = test["user"].unique()

    recommendations = []

    for user in users:

        rated = train[train["user"] == user]["item"]
        candidates = np.setdiff1d(all_items, rated)

        scores = mf.predict_for_user(user, candidates)
        scores = scores.dropna()

        top10 = scores.sort_values(ascending=False).head(10)

        for item, score in top10.items():
            recommendations.append({
                "user": user,
                "item": item,
                "score": score
            })

    recs = pd.DataFrame(recommendations)


    truth = test[test["rating"] >= 4]
    metrics = rla.compute(recs, truth)

    all_metrics.append(metrics.mean())

    print("\nRanking Metrics:")
    print(metrics.mean())



print("\n========================")
print("FINAL CROSS VALIDATION RESULTS")
print("========================")

print("\nRMSE (mean):", np.mean(rmse_list))
print("MAE (mean):", np.mean(mae_list))

print("\nRanking Metrics (mean over folds):")
print(pd.DataFrame(all_metrics).mean())
````

## d) 

````python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import seaborn as sns



print("\n--- Visualizing Embeddings ---")


user_factors = mf.user_features_
item_factors = mf.item_features_

user_idx = mf.user_index_
item_idx = mf.item_index_

user_df = pd.DataFrame(user_factors, index=user_idx)
item_df = pd.DataFrame(item_factors, index=item_idx)


pca = PCA(n_components=2)
pca.fit(item_factors)

user_2d = pca.transform(user_factors)
item_2d = pca.transform(item_factors)

user_2d_df = pd.DataFrame(user_2d, index=user_idx, columns=['PC1', 'PC2'])
item_2d_df = pd.DataFrame(item_2d, index=item_idx, columns=['PC1', 'PC2'])


target_user = users[0]

user_history = train[train['user'] == target_user]
highly_rated = user_history[user_history['rating'] >= 4.0]['item'].tolist()

user_recs = recs[recs['user'] == target_user]['item'].tolist()

highly_rated = [i for i in highly_rated if i in item_2d_df.index]
user_recs = [i for i in user_recs if i in item_2d_df.index]

plt.figure(figsize=(12, 8))

plt.scatter(item_2d_df['PC1'], item_2d_df['PC2'], color='lightgray', alpha=0.3, label='All Items')

plt.scatter(item_2d_df.loc[highly_rated, 'PC1'], item_2d_df.loc[highly_rated, 'PC2'], 
            color='royalblue', s=70, label='Highly Rated (Truth)', edgecolor='black')

plt.scatter(item_2d_df.loc[user_recs, 'PC1'], item_2d_df.loc[user_recs, 'PC2'], 
            color='limegreen', s=70, marker='s', label='Recommended (Top 10)', edgecolor='black')

plt.scatter(user_2d_df.loc[target_user, 'PC1'], user_2d_df.loc[target_user, 'PC2'], 
            color='crimson', s=250, marker='*', label=f'User {target_user}', edgecolor='black')

variance_explained = pca.explained_variance_ratio_
plt.title(f'2D PCA Projection of Embeddings (User {target_user})', fontsize=14)
plt.xlabel(f'Principal Component 1 ({variance_explained[0]:.2%} Explained Variance)')
plt.ylabel(f'Principal Component 2 ({variance_explained[1]:.2%} Explained Variance)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

print(f"Total Explained Variance by 2D projection: {variance_explained.sum():.2%}")
````

Here we have extracted the user as well as the item features and have reduced the dimension with PCA. 

**Interpretation**   
With the PCA we have recued the dimensions from 50 (this was the features hyperparameter) to two dimensions. This can or cannot align with reality. 
If there are two dimensions which are very important e.g. "is this film rated >4 stars" then the result of the PCA can align with reality quite good. 
In our case we could somehow draw a circle around all green and red dots, but this would incorporate probably 1/3 of all films. Meaning this doesn't really 
show which films are interesting to the user. 
Furthermore if we would select the 10 films with the lowest euclidean distance to the red star we wouldn't have found a single movie the system wants to recommend to the user. 

We can also see from the components that they explain 4.67 and 3.45 % of the variance. This means that we can only "see about 8% of the variance" which can 
be vaguely interpreted as we only see 1/10 of the deciding factors. This is similar to only looking at a actor in a supporting role and according to him or her recommend the film to a user.
